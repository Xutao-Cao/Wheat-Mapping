```python
import torch
import numpy as np
import sys
# module_path="D:\\Tony\\SRTP\\WheatMappingModel"
sys.path.append("D:\\Tony\\SRTP\\WheatMappingModel") 
from logger import PrettyLogger
from helper import WMMhelper
```

```python
logger = PrettyLogger()
helper = WMMhelper()
DEVICE = torch.device("cuda:0")
```

```python
x_train = helper.input_x(["C:\\Users\\hp\\x1.npy","C:\\Users\\hp\\x2.npy","C:\\Users\\hp\\x3.npy","C:\\Users\\hp\\x4.npy"])
y_train = helper.input_y(["C:\\Users\\hp\\y1.npy","C:\\Users\\hp\\y2.npy","C:\\Users\\hp\\y3.npy","C:\\Users\\hp\\y4.npy"])
x_test = helper.input_x(["C:\\Users\\hp\\x5.npy","C:\\Users\\hp\\x6.npy"])
y_test = helper.input_y(["C:\\Users\\hp\\y5.npy","C:\\Users\\hp\\y6.npy"])
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
# print(x_train)
```

```python
train_dataloader = helper.make_dataloader(x_train, y_train)
test_dataloader = helper.make_dataloader(x_test, y_test)
print(train_dataloader)
print(len(train_dataloader.dataset))
```

```python
for i, batch in enumerate(train_dataloader):
    xt_train_batch = batch["x"].to(DEVICE)
    yt_train_batch = batch["y"].to(DEVICE)
    print(xt_train_batch.size())
    print(yt_train_batch.size())
for i, batch in enumerate(test_dataloader):
    xt_test_batch = batch["x"].to(DEVICE)
    yt_test_batch = batch["y"].to(DEVICE)
    print(xt_test_batch.size())
    print(yt_test_batch.size())
```

```python
net = helper.build_model()
net.to(DEVICE)
helper.train_model(net, train_dataloader, test_dataloader, DEVICE, logger)
```
